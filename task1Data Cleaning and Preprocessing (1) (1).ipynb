```python
import pandas as pd
```

```python
#Load the dataset
df = pd.read_csv("C:/Users/Admin/Desktop/Customer_Personality_Analysis.csv")
```

```python
df.head()
```

```python
df.shape
```

```python
#Checking Missing Values
df.isnull()
```

```python
#Remove Duplicate Rows
df = df.drop_duplicates()
```

```python
#View Cleaned Data
print(df)
```

```python
#Standardize Text Values
df['Marital_Status'] = df['Marital_Status'].str.lower()
```

```python
#Standardize Text Values
df['Education'] = df['Education'].str.lower()
```

```python
#Conversion of Date Format
df['Dt_Customer'] = pd.to_datetime(df['Dt_Customer'], dayfirst=True)
```

```python
# Rename column headers to be clean and uniform (e.g., lowercase, no spaces).
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
```

```python
print(df)
```

```python
df['income'] = df['income'].astype(float)
```

```python
df.to_csv("cleaned_dataset.csv", index=False)
```
